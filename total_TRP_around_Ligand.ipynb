```python
import MDAnalysis as mda
import numpy as np
import pandas as pd
import warnings

# Suppress warnings for cleaner output
warnings.simplefilter("ignore", category=np.VisibleDeprecationWarning)

# File paths
TPR = 'md_0_sim_resp_24trp-4mnt_500ns.tpr' 
XTC = 'md_0_sim_resp_24trp-4mnt_500ns_pbc_center.xtc'
# Load the universe
u = mda.Universe(TPR, XTC)

# Atom selections
trp = u.select_atoms("resname TRP")
ta = u.select_atoms("resname MNT")  # Select all TA5 molecules

# Analysis parameters
cutoff = 12.0  # Ångström
start_frame = 5000
trp_counts = []

# Main analysis loop
for ts in u.trajectory[start_frame:]:
    # Update positions with unwrapped coordinates
    trp_coms = trp.center_of_mass(unwrap=True, compound="residues")
    ta_coms = ta.center_of_mass(unwrap=True, compound="residues")  # n TA5 molecules

    # Calculate distances of each TRP to all TA5 molecules
    dist_matrix = np.linalg.norm(trp_coms[:, np.newaxis] - ta_coms, axis=2)  # Shape (num_TRP, num_TA5)
    
    # Find the minimum distance to any TA5 for each TRP
    min_distances = np.min(dist_matrix, axis=1)
    
    # Count TRPs within the cutoff distance from any TA5
    nearby_trps = np.sum(min_distances <= cutoff)
    trp_counts.append(nearby_trps)

# Data processing
results = pd.DataFrame({'TRP_count': trp_counts})
average_trp = results.mean().values[0]

print(f"Average number of TRP residues around TA5 (n = {len(ta)}): {average_trp:.2f}")
```

```python
import MDAnalysis as mda
import numpy as np
import pandas as pd
import warnings

# Suppress warnings for cleaner output
warnings.simplefilter("ignore", category=np.VisibleDeprecationWarning)

TPR = 'md_0_sim_resp_24trp-4mnt_500ns.tpr' 
XTC = 'md_0_sim_resp_24trp-4mnt_500ns_pbc_center.xtc'

# Load the universe
u = mda.Universe(TPR, XTC)

# Atom selections
trp = u.select_atoms("resname TRP")
ta = u.select_atoms("resname MNT")  # Select all TA5 molecules

# Analysis parameters
cutoff = 12.0  # Ångström
start_frame = 5000
trp_counts = []

# Main analysis loop
for ts in u.trajectory[start_frame:]:
    # Update positions with unwrapped coordinates
    trp_coms = trp.center_of_mass(unwrap=True, compound="residues")
    ta_coms = ta.center_of_mass(unwrap=True, compound="residues")  # All TA5 molecules

    total_trp_count = 0  # Initialize counter for each frame

    # Iterate over all TA5 molecules
    for ta_com in ta_coms:
        distances = np.linalg.norm(trp_coms - ta_com, axis=1)  # Compute distances for all TRPs
        trp_count = np.sum(distances <= cutoff)  # Count TRPs within cutoff for this TA5
        total_trp_count += trp_count  # Accumulate counts from all TA5 molecules

    trp_counts.append(total_trp_count)  # Store total TRP count for this frame

# Data processing
results = pd.DataFrame({'TRP_count': trp_counts})
average_trp = results.mean().values[0]

# Print the number of TA5 dynamically and the result
print(f"Average number of TRP residues around TA5 (n = {len(ta_coms)}): {average_trp:.2f}")
```

```python
import MDAnalysis as mda
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from tqdm import tqdm

# # Load trajectory
# TPR = 'md_0_sim_sim_resp_24trp-1ta_500ns.tpr'
# XTC = 'md_0_sim_resp_24trp-1ta_500ns_pbc_center.xtc'
TPR = 'md_0_sim_resp_24trp-2ta_500ns.tpr'
XTC = 'md_0_sim_resp_24trp-2ta_500ns_pbc_center.xtc'
u = mda.Universe(TPR, XTC)

# Select atoms
trp = u.select_atoms("resname TRP")
ta = u.select_atoms("resname TA5")

# Define cutoff distances
CUTOFF = 15.0
cutoff_cluster = 7.5

# Initialize lists for storing results
free_trp_counts = []
max_cluster_sizes = []
oligomer_counts = []
time_ns = []

# Process trajectory
for ts in tqdm(u.trajectory[5000:]):
    # Calculate centers of mass
    trp_coms = trp.center_of_mass(unwrap=True, compound="residues")
    ta_coms = ta.center_of_mass(unwrap=True, compound="residues")
    
    # Identify free TRPs (outside all TA5 cutoffs)
    free_indices = []
    for i, t_com in enumerate(trp_coms):
        if all(np.linalg.norm(t_com - ta_com) >= CUTOFF for ta_com in ta_coms):
            free_indices.append(i)
    
    free_trp_counts.append(len(free_indices))
    
    # Skip RDF calculation if <2 free TRPs
    if len(free_indices) < 2:
        continue

    free_positions = trp_coms[free_indices]
    
    # Calculate pairwise distances
    distances = []
    n = len(free_positions)
    for i in range(n - 1):
        for j in range(i + 1, n):
            distances.append(np.linalg.norm(free_positions[i] - free_positions[j]))

    # Cluster analysis
    clusters = []
    visited = set()
    
    for i in range(n):
        if i in visited:
            continue
        cluster = {i}
        to_check = {i}
        
        while to_check:
            idx = to_check.pop()
            visited.add(idx)
            for j in range(n):
                if j not in visited and np.linalg.norm(free_positions[idx] - free_positions[j]) < cutoff_cluster:
                    cluster.add(j)
                    to_check.add(j)
        
        clusters.append(cluster)

    max_cluster_size = max(len(cluster) for cluster in clusters) if clusters else 0
    oligomer_count = len([c for c in clusters if len(c) > 1])
    
    max_cluster_sizes.append(max_cluster_size)
    oligomer_counts.append(oligomer_count)
    time_ns.append(ts.time / 1000)

# Plot results
plt.figure(figsize=(8, 6))
plt.plot(time_ns, max_cluster_sizes, color='#2c7bb6', lw=1.2)
plt.xlabel('Simulation Time (ns)', fontweight='bold', fontsize=12)
plt.ylabel('Maximum Cluster Size', fontweight='bold', fontsize=12)
plt.title('TRP Oligomerization Dynamics Near Tannic Acid', fontsize=14)
plt.grid(alpha=0.3, linestyle='--')
plt.xlim(250, 500)
plt.ylim(0, max(max_cluster_sizes) + 1)
plt.tight_layout()
plt.show()
```

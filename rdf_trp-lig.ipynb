```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# Set publication-style parameters for optimal presentation
plt.rcParams.update({
    "font.family": "Times New Roman",  # Preferred serif font for publications
    "font.size": 35,                   # General font size
    "axes.labelsize": 35,              # Increased axis label size
    "axes.titlesize": 20,              # Title size
    "axes.labelweight": 'bold',        # Bold labels for emphasis
    "axes.linewidth": 3,               # Thicker axis lines for clarity
    "lines.linewidth": 3,              # Thicker lines for visibility
    "xtick.labelsize": 30,             # X-axis tick label size
    "ytick.labelsize": 25,             # Y-axis tick label size
    "xtick.direction": 'in',           # Tick marks directed inward for clean layout
    "ytick.direction": 'in',           # Tick marks directed inward for clean layout
    "xtick.top": True,                 # Display ticks on top
    "ytick.right": True,               # Display ticks on the right
    "xtick.major.size": 10,            # Major tick size for clarity
    "ytick.major.size": 10,            # Major tick size for clarity
    "xtick.major.width": 3,            # Major tick width
    "ytick.major.width": 3,            # Major tick width
    "xtick.minor.size": 5,             # Minor tick size for precision
    "ytick.minor.size": 5,             # Minor tick size for precision
    "xtick.minor.width": 3,            # Minor tick width
    "ytick.minor.width": 3,            # Minor tick width
    "xtick.minor.visible": True,       # Show minor ticks
    "ytick.minor.visible": True,       # Show minor ticks
    "legend.fontsize": 15,             # Larger legend font size for better readability
    "legend.frameon": True,            # Frame around the legend for clarity
    "legend.loc": 'upper right',       # Legend placement (adjustable)
    "legend.markerscale": 1.5,         # Adjust marker size in the legend
    "legend.borderpad": 1.0,           # Padding around the legend
    "savefig.dpi": 500,                # High DPI for publication-quality images
    "savefig.bbox": 'tight',           # Tight bounding box for saving
})

# List of xvg files
xvg_files = [
    "rdf_ta-trp.xvg", "rdf_1sbt-trp.xvg", "rdf_1mnt-trp.xvg",
    "rdf_2sbt-trp.xvg", "rdf_2mnt-trp.xvg", "rdf_4sbt-trp.xvg",
    "rdf_4mnt-trp.xvg"
]

# Custom legend labels
legend_labels = [
    "S1 (1 TA)", "S2 (1 SBT)", "S3 (1 MNT)",
    "S4 (2 SBT)", "S5 (2 MNT)", "S6 (4 SBT)",
    "S7 (4 MNT)"
]

# Function to read xvg files and extract data
def read_xvg(filename):
    x, y = [], []
    with open(filename, "r") as f:
        for line in f:
            if line.startswith(("#", "@")):
                continue  # Skip comments and metadata
            values = line.split()
            x.append(float(values[0]) * 10)  # Convert nm to Å
            y.append(float(values[1]))
    return np.array(x), np.array(y)

# Plot data
plt.figure(figsize=(10, 8))  # Adjusted figure size
colors = plt.cm.viridis(np.linspace(0, 1, len(xvg_files)))  # Colormap for better distinction

for file, label, color in zip(xvg_files, legend_labels, colors):
    x, y = read_xvg(file)
    plt.plot(x, y, label=label, color=color)  # Use custom legend labels

# Formatting the plot
# plt.xlabel(r"$r$ (Å)", fontsize=40, fontweight="bold")  # Increased axis label size
# plt.ylabel(r"$g(r)$", fontsize=40, fontweight="bold")  # Increased axis label size
plt.xlabel("r(Å)", fontsize=30)
plt.ylabel("g(r)", fontsize=30)
plt.xticks(fontsize=30)  # Adjusted tick label size
plt.yticks(fontsize=30)  # Adjusted tick label size
plt.legend(ncol=1, fontsize=18, frameon=True, fancybox=False, shadow=True, loc='upper right', prop={'weight': 'normal'})  
plt.tight_layout()  # Optimize layout
plt.xlim(0, 25)
plt.savefig('rdf_trp-lig.png', dpi=500, bbox_inches="tight")
# Show plot
plt.show()
```

```python
import MDAnalysis as mda
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  
import pandas as pd
from tqdm import tqdm  

TPR = 'md_0_sim_resp_24trp-1ta_500ns.tpr' 
XTC = 'md_0_sim_resp_24trp-1ta_500ns_pbc_center.xtc'

u = mda.Universe(TPR, XTC)

trp = u.select_atoms("resname TRP")
ta = u.select_atoms("resname TA5")

cutoff_1 = 17.0

trp_counts_per_frame = []

for ts in tqdm(u.trajectory[5000:]):  
    trp_coms = trp.center_of_mass(unwrap=True, compound="fragments")
    ta_coms = ta.center_of_mass(unwrap=True, compound="fragments")
    
    total_trp_count = 0 
    for ta_com in ta_coms:
        distances = [np.linalg.norm(trp_com - ta_com) for trp_com in trp_coms]
        trp_count = np.sum(np.array(distances) <= cutoff_1)
        total_trp_count += trp_count
    
    trp_counts_per_frame.append(total_trp_count)

data1 = pd.DataFrame(trp_counts_per_frame, columns=['Total_TRP_Count'])

# Calculate average of total interactions (exclude zeros if needed)
data_ta = data1[data1 != 0]
average1 = data_ta.mean(axis=0)
print(average1)
```
